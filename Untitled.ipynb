```python
import pandas as pd

df = pd.read_csv ('job_skills.csv', encoding='latin-1')
pd.set_option('display.max_columns', None)
df.head(500)
```

```python
df['Category'] = df['Category'].str.replace('Program Management',' Management')
df['Category'] = df['Category'].str.replace('Partnerships',' Management')
df['Category'] = df['Category'].str.replace('People Operations',' Management')

df['Category'] = df['Category'].str.replace('Sales & Account Management ',' Business Strategy')
df['Category'] = df['Category'].str.replace('Finance',' Business Strategy')
df['Category'] = df['Category'].str.replace('Sales Operations',' Business Strategy')

df['Category'] = df['Category'].str.replace('Real Estate & Workplace Service','Manufacturing & Supply Chain')
df['Category'] = df['Category'].str.replace('Legal & Government Relations','Administrative')
df['Category'] = df['Category'].str.replace('Data Center & Network  ','Network Engineering ')
df['Category'] = df['Category'].str.replace('Technical Infrastructure   ','Technical Solutions   ')
df['Category'] = df['Category'].str.replace('Product & Customer Support   ','Marketing & Communications   ')

df = df[pd.notnull(df['Responsibilities'])]

df['Loc'] = df['Location'].apply(lambda x:" ".join(x.split()[-1:]))


df['BA'] = df['Minimum Qualifications'].apply(lambda x:1 if "BA" in x else 0)
df['BS'] = df['Minimum Qualifications'].apply(lambda x:1 if "BS" in x else 0)
df['MS'] = df['Minimum Qualifications'].apply(lambda x:1 if "MS" in x else 0)
df["Bachelor's degree"] = df['Minimum Qualifications'].apply(lambda x:1 if "Bachelor's degree" in x else 0)

df["Degree in CS"] = df['Minimum Qualifications'].apply(lambda x:1 if "Bachelor's degree in Computer Science" in x else 0)
df["Degree in Electrical Engineering"] = df['Minimum Qualifications'].apply(lambda x:1 if "degree in Electrical Engineering" in x else 0)

df['Loc'] = df['Loc'].str.replace('Product & Customer Support   ','Marketing & Communications   ')



def country_reigon(inpt):
    if inpt == 'Singapore' or inpt == 'Thailand' or inpt == 'China' or inpt == 'Taiwan' or inpt == 'India'or inpt == 'Japan'or inpt == 'Philippines'or inpt == 'Indonesia'or inpt == 'Korea'or inpt == 'Russia'or inpt == 'LG'or inpt == 'Kong':
        return 'Asia'
    
    if  inpt == 'Poland' or  inpt == 'Croatia' or inpt =='Sweden' or inpt =='Netherlands' or inpt =='Israel' or inpt == 'Ireland' or inpt == 'Kingdom' or inpt == 'Germany' or inpt == 'Switzerland'or inpt == 'France'or inpt =='Italy' or inpt =='Norway' or inpt =='Turkey'  or inpt =='Finland'   or inpt =='Belgium' or inpt =='Romania'or inpt =='Austria' or inpt =='Denmark' or inpt =='Czechia' or inpt =='Hungary' or inpt =='Spain' or inpt =='Slovakia' or inpt =='Lithuania' or inpt =='Ukraine' or inpt =='Portugal' or inpt =='Greece' :
        return 'Europe'
    if  inpt == 'Brazil' or  inpt =='States' or inpt =='Mexico' or inpt =='Canada' or inpt =='Argentina' or inpt == 'Colombia' or inpt == 'USA' :
        return 'America'
    if  inpt == 'Emirates' or inpt =='Nigeria' or inpt =='Kenya' :
        return 'Africa'
    else:
        return inpt
    
df['Loc'] = df['Loc'].apply(country_reigon)

def title(inpt):
    if inpt == 'MBA Intern, Summer 2018' or inpt == 'MBA Intern 2018' or inpt == 'BOLD Intern, Summer 2018' or inpt == 'Business Intern 2018' :
        return 'Business Intern'
    
 
    else:
        return inpt
df['Title'] = df['Title'].apply(title)
```

```python
df = df.drop(columns=['Location','Responsibilities', 'Minimum Qualifications', 'Preferred Qualifications'])
df
            
```

```python
df = pd.get_dummies(df)
df.head(5)
```

```python
df.shape
```

```python

X=df.drop(['Category_ Business Strategy','Category_ Management','Category_Administrative','Category_Business Strategy','Category_Data Center & Network','Category_Developer Relations','Category_Hardware Engineering','Category_IT & Data Management','Category_Manufacturing & Supply Chain','Category_Manufacturing & Supply Chains','Category_Marketing & Communications','Category_Network Engineering','Category_Product & Customer Support','Category_Sales & Account Management','Category_Software Engineering','Category_Technical Infrastructure','Category_Technical Solutions','Category_Technical Writing','Category_User Experience & Design'],axis=1)
y = df[['Category_ Business Strategy','Category_ Management','Category_Administrative','Category_Business Strategy','Category_Data Center & Network','Category_Developer Relations','Category_Hardware Engineering','Category_IT & Data Management','Category_Manufacturing & Supply Chain','Category_Manufacturing & Supply Chains','Category_Marketing & Communications','Category_Network Engineering','Category_Product & Customer Support','Category_Sales & Account Management','Category_Software Engineering','Category_Technical Infrastructure','Category_Technical Solutions','Category_Technical Writing','Category_User Experience & Design']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10, 50, 100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(X_train, y_train)

best_model = grid_fit.best_estimator_

best_model.score(X_test, y_test)
```
